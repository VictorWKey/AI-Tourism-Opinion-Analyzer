# An√°lisis de Sentimientos de Opiniones Tur√≠sticas - Todas las Ciudades

Este notebook realiza un an√°lisis comparativo de sentimientos sobre las opiniones tur√≠sticas recopiladas de **todas las ciudades**, utilizando **tres m√©todos diferentes** para evaluar la percepci√≥n de los visitantes sobre diferentes atracciones tur√≠sticas.

## üéØ Objetivos del An√°lisis

### An√°lisis Principal:
- Cargar y procesar dataset consolidado de todas las ciudades
- Realizar an√°lisis de sentimientos mediante tres enfoques complementarios
- Comparar y evaluar la concordancia entre los diferentes m√©todos
- Generar visualizaciones comparativas y estad√≠sticas descriptivas
- Exportar resultados para an√°lisis posteriores

### M√©todos de An√°lisis de Sentimientos:

#### 1Ô∏è‚É£ **M√©todo Basado en Calificaciones** 
- Mapeo directo de calificaciones de estrellas (1-5) a sentimientos
- **Mapeo:** 4-5 estrellas ‚Üí Positivo | 3 estrellas ‚Üí Neutro | 1-2 estrellas ‚Üí Negativo
- **Ventaja:** Simplicidad y interpretabilidad directa

#### 2Ô∏è‚É£ **Modelo HuggingFace - BERT Multiling√ºe**
- Modelo: `nlptown/bert-base-multilingual-uncased-sentiment`
- **Caracter√≠sticas:** An√°lisis basado en contenido textual, soporte multiling√ºe
- **Especializaci√≥n:** Sentimientos generales en rese√±as y opiniones

#### 3Ô∏è‚É£ **Modelo Cardiff NLP - Twitter XLM-RoBERTa**
- Modelo: `cardiffnlp/twitter-xlm-roberta-base-sentiment`
- **Caracter√≠sticas:** Optimizado para texto corto y expresivo
- **Especializaci√≥n:** Captura de emociones en textos breves (entrenado en Twitter)

## üìä Estructura del An√°lisis

1. **Configuraci√≥n y Carga de Datos**
2. **An√°lisis Base por Calificaciones**
3. **An√°lisis con Modelo HuggingFace**
4. **An√°lisis con Modelo Cardiff NLP**
5. **Comparaciones y Evaluaci√≥n de Concordancia**
6. **Visualizaciones y Resultados**
7. **Exportaci√≥n de Datos Procesados**

## üîç Resultados Esperados
- Identificaci√≥n de patrones de sentimientos por m√©todo y por ciudad
- An√°lisis de concordancias y discrepancias entre enfoques
- Evaluaci√≥n de la efectividad de cada modelo en contexto tur√≠stico
- Dataset consolidado enriquecido con m√∫ltiples clasificaciones de sentimientos

## 5Ô∏è‚É£ An√°lisis Comparativo Global

En esta secci√≥n realizamos un an√°lisis comparativo comprehensivo de los tres m√©todos evaluados, proporcionando una visi√≥n global de su rendimiento y concordancia.

```python
ciudades = df_con_cardiff['Ciudad'].unique()

# Calcular concordancias entre pares y concordancia perfecta entre los tres m√©todos de forma vectorizada
concordancias = {
    "cal_hf": (df_con_cardiff['SentimientoPorCalificacion'] == df_con_cardiff['SentimientoPorHF']),
    "cal_cardiff": (df_con_cardiff['SentimientoPorCalificacion'] == df_con_cardiff['SentimientoPorCardiff']),
    "hf_cardiff": (df_con_cardiff['SentimientoPorHF'] == df_con_cardiff['SentimientoPorCardiff']),
    "total_tres": (df_con_cardiff['SentimientoPorCalificacion'] == df_con_cardiff['SentimientoPorHF']) & 
                   (df_con_cardiff['SentimientoPorCalificacion'] == df_con_cardiff['SentimientoPorCardiff'])
}

# Calcular porcentajes de concordancia
resultados_concordancia = {k: v.mean() * 100 for k, v in concordancias.items()}

# Resumen final
print(f"üìä RESUMEN FINAL - Todas las ciudades")
print(f"‚úÖ Opiniones analizadas: {len(df_con_cardiff)} | Atracciones: {df_con_cardiff['Atraccion'].nunique()}")
print(f"üèôÔ∏è Ciudades: {', '.join(ciudades)}")
print(f"\nüîç Concordancias entre pares:")
print(f"   üéØ Calificaci√≥n vs HuggingFace: {resultados_concordancia['cal_hf']:.1f}%")
print(f"   üéØ Calificaci√≥n vs Cardiff NLP: {resultados_concordancia['cal_cardiff']:.1f}%")
print(f"   üéØ HuggingFace vs Cardiff NLP: {resultados_concordancia['hf_cardiff']:.1f}%")
print(f"\n‚ú® Concordancia perfecta (3 m√©todos): {resultados_concordancia['total_tres']:.1f}%")
```
